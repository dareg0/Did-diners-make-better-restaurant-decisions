```python
import pandas as pd
```

```python
df_yelp = pd.read_csv('Data/YELP_ind_CAMIS_merged.csv')
df_insp = pd.read_csv('Data/most_recent_inspections.csv')
```

```python
# clean yelp data
df_yelp.info()
```

```python
df_yelp.head()
```

```python
df_yelp.isna().sum()   
```

```python
df_insp['CAMIS'].nunique()
```

```python
# replace yelp Price "$" to 1,2,3,4 score
df_yelp.loc[df_yelp['Price'] == "$", 'Price'] = 1
df_yelp.loc[df_yelp['Price'] == "$$", 'Price'] = 2
df_yelp.loc[df_yelp['Price'] == "$$$", 'Price'] = 3
df_yelp.loc[df_yelp['Price'] == "$$$$", 'Price'] = 4

# replace some weired signs 
df_yelp.loc[df_yelp['Price'] == "€€€", 'Price'] = 3
df_yelp.loc[df_yelp['Price'] == "££££", 'Price'] = 4
df_yelp.loc[df_yelp['Price'] == "££", 'Price'] = 2
df_yelp.loc[df_yelp['Price'] == "£", 'Price'] = 1

# but still 2574 na values
```

```python
df_insp.info()
```

```python
df_insp.head()
```

```python
# join two dataframes on CAMIS column index
df = pd.concat([df_yelp.set_index('CAMIS'),df_insp.set_index('CAMIS')], axis=1, join='inner').reset_index()
```

```python
df.info()
```

```python
# drop NA Values
df = df[pd.notnull(df['Price'])]
df = df[pd.notnull(df['Reviews'])]
df = df[pd.notnull(df['Categories'])]
```

```python
df.to_csv('merged_yelp_inspections.csv', index=False)
```
